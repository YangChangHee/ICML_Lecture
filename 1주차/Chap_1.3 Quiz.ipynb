# 다차원 회귀

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

# Data 1500개

```python
num_data=1500
x1=np.random.rand(num_data)*10
```

```python
x2=np.random.rand(num_data)*10
```

```python
X=np.array([x1,x2]).T
```

# Log, Cos의 분류

```python
y=np.log(x1)+np.cos(x2)
```

```python
train_X, train_y = X[:1000,:],y[:1000]
val_X, val_y=X[1000:12000,:],y[1000:12000]
test_X,test_y=X[1200:,:],y[1200:]
```

```python
fig=plt.figure(figsize=(12,5))
ax1 = fig.add_subplot(1,3,1,projection="3d")
ax1.scatter(train_X[:,0], train_X[:,1], train_y, c=train_y, cmap='jet')

ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
ax1.set_zlabel('y')
ax1.set_title('Train Set Distribution')
ax1.set_zlim(-10, 6)
ax1.view_init(0, 90)
ax1.invert_xaxis()
ax2 = fig.add_subplot(1, 3, 2, projection='3d')
ax2.scatter(val_X[:, 0], val_X[:, 1], val_y, c=val_y, cmap='jet')

ax2.set_xlabel('x1')
ax2.set_ylabel('x2')
ax2.set_zlabel('y')
ax2.set_title('Validation Set Distribution')
ax2.set_zlim(-10, 6)
ax2.view_init(0, 0)
ax2.invert_xaxis()
ax3 = fig.add_subplot(1, 3, 3, projection='3d')
ax3.scatter(test_X[:, 0], test_X[:, 1], test_y, c=test_y, cmap='jet')

ax3.set_xlabel('x1')
ax3.set_ylabel('x2')
ax3.set_zlabel('y')
ax3.set_title('Test Set Distribution')
ax3.set_zlim(-10, 6)
ax3.view_init(40, -60)
ax3.invert_xaxis()
plt.show()
```
